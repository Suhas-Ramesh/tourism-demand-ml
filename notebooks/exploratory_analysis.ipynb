# Exploratory Data Analysis (EDA)

This notebook explores tourism demand data across multiple countries.

The goal of this analysis is to understand the structure of the dataset, identify patterns, and visualize relationships between tourism indicators before building machine learning models.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
data = pd.read_csv("../data/tourism_dataset.csv")

data.head()
```

```python
# Dataset information
data.info()

# Summary statistics
data.describe()
```

```python
# Missing value analysis
data.isnull().sum()
```

```python
# Distribution of numerical variables
data.hist(figsize=(12,10))

plt.tight_layout()
plt.show()
```

```python
# Correlation analysis
plt.figure(figsize=(12,8))

sns.heatmap(data.corr(), cmap="coolwarm")

plt.title("Correlation Matrix")

plt.show()
```

```python
# Tourism demand trend over time
sns.lineplot(
    data=data,
    x="Year",
    y="Inbound-Total Arrival(overnight stay)( In thousands)",
    hue="Country"
)

plt.title("Tourism Demand Over Time")

plt.show()
```
